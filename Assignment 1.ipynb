```python
plt.figure(figsize=(11, 6))
df.groupby(by="brand")["price"].mean().plot(kind='bar')
plt.title("Average resale value of cars brand-wise")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=75)
sns.boxplot(x='brand', y='price', data=df)
```

```python
df["fuel_type"].value_counts().plot(kind='pie', autopct="%1.2f%%")
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df["engine_capacity(CC)"])
```

```python
sns.boxplot(x='transmission', y='price', data=df)
```

```python
sns.kdeplot(data=df, x='price', hue='fuel_type', fill=True)
```

```python
df['reg_number'].str[:2].value_counts().plot(kind='pie')
plt.title("Distribution of cars by state")
```

```python
plt.plot(df.groupby(by="make_year")['price'].mean())
plt.title("Average car price by year")
plt.xlabel("Year")
plt.ylabel("Avg. price")
```

Inferences:
1. Data is quite cleaned so we can directly make visualisation from it.
2. We can clearly see that Brands like Mahindra, Hyundai, and Skoda show a wider IQR, indicating greater price variation in their car offerings from the median value.
3. Fuel type: Petrol is the most widely used one and electric cars are comparatively very less.
4. Prices for cars with 'automatic' transmission type is more than that with 'manual' transmission type.
5. Average resale value of cars of the brand "MG" is the highest among all.
6. Cars of brand "Maruti" are in the maximum number for resale.
7. Cars registered in Karnataka state are available in the maximum number for resale.
8. Average car price increased over the years with a dip just shown around the year 2023.
9. There is negative correlation between 'KM driven' and 'price' which indicates price goes down if the car is driven more which is quite obvious.
10. We can clearly see that mean price for 1st owner cars>2nd owner cars>3rd owner cars as the price should decrease with hands increasing.
